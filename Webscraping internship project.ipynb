```python
soup=BeautifulSoup(page.content)
soup
```

```python
header=soup.find_all(["h1","h2","h3","h4","h5","h6"])
header
```

```python
import pandas as pd
page=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
loc=[]
for i in soup.find_all("h3",class_="lister-item-header"):
    loc.append(i.text.replace("\n",""))
loc
```

```python
rating=[]
for i in soup.find_all("div",class_="inline-block ratings-imdb-rating"):
    rating.append(i.text.replace("\n",""))
rating
```

```python
df=pd.DataFrame({"Movie":loc,"Rating":rating})
df
```

```python
page=requests.get("https://www.imdb.com/list/ls009997493/")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
movie=[]
for i in soup.find_all("h3",class_="lister-item-header"):
    movie.append(i.text.replace("\n",""))
movie
```

```python
Rating=[]
for i in soup.find_all("div",class_="ipl-rating-star small"):
    Rating.append(i.text.replace("\n",""))
Rating
```

```python
df=pd.DataFrame({"movie":movie,"rating":Rating})
df
```
